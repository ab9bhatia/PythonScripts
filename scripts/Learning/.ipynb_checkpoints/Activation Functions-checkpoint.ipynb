```python
from matplotlib.mlab import normpdf
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
def sigmoid_fun(scale,step,label=None):
    x = np.arange(-scale, scale, step)
    y = 1 / (1+np.exp(-x))
    plt.plot(x,y,label=label)
    plt.grid(True)
    plt.legend(loc=0)
    #plt.show()
```

```python
sigmoid_fun(10,0.01,label='Sigmoid')
plt.show()
```

```python
def softmax_fun(scale,step,label=None):
    x = np.arange(-scale, scale, step)
    y = np.exp(x) / np.sum(np.exp(x))  
    plt.plot(x,y,label=label)
    plt.grid(True)
    plt.legend(loc=0)
    #plt.show()
```

```python
softmax_fun(10,0.01,label='Softmax')
plt.show()
```

```python
def relu(scale,step,label=None):
    x = np.arange(-scale, scale, step)
    y = np.array([np.maximum(0, x[i]) for i in np.arange(len(x))])
    plt.plot(x,y,label=label)
    plt.grid(True)
    plt.legend(loc=0)
    #plt.show()
```

```python
relu(10,0.01,label='ReLu')
plt.show()
```

```python
def tanh_fun(scale,step,label=None):
    x = np.arange(-scale, scale, step)
    y = (2 / (1+np.exp(-2*x)) - 1)
    plt.plot(x,y,label=label)
    plt.grid(True)
    plt.legend(loc=0)
    #plt.show()
```

```python
tanh_fun(10,0.01,label='Tan H')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sigmoid_fun(10,0.01,label='Sigmoid')
softmax_fun(10,0.01,label='Softmax')
relu(3,0.01,label='ReLu')
tanh_fun(10,0.01,label='Tan H')
```
