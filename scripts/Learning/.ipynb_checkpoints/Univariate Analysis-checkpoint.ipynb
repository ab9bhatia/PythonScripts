```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sachin = pd.read_csv('./data/tendulkar_ODI.csv')
print(sachin.shape)
sachin.head()
```

```python
# Remove '*' from runs
# Remove rows where runs ='DNB'

remove_filter = (sachin['Runs'].isin(['DNB','TDNB']))
update_filter = (sachin['Runs'].str.endswith('*'))
sachin.drop(labels = sachin[remove_filter].index,inplace = True)
sachin.loc[update_filter,'Runs'] = sachin.loc[update_filter,'Runs'].str.replace('*','')
sachin = sachin.sort_values(by = 'Runs')
```

#### Cast str to int

```python
sachin['Runs'] = sachin['Runs'].apply(lambda x: int(x))
sachin['4s'] = sachin['4s'].apply(lambda x: int(x))
```

#### Binning(convert numeric variables into ordered categorical) of Run scores

```python
plt.figure(figsize=(14,7))
x = sachin['Runs']
bins = np.arange(np.min(sachin['Runs']),np.max(sachin['Runs']),10)

plt.hist(x = x, bins=bins, rwidth=.9)
plt.xticks(np.arange(np.min(sachin['Runs']),np.max(sachin['Runs']),10))
plt.grid()
    
plt.show()
```

#### Number of fours

```python
plt.figure(figsize=(14,7))
x = sachin['4s']
bins = np.arange(np.min(sachin['4s']),np.max(sachin['4s']),1)

plt.hist(x = x, bins=bins, rwidth=.9)
plt.xticks(np.arange(np.min(sachin['4s']),np.max(sachin['4s']),1))
plt.grid()
    
plt.show()
```

```python
from scipy import stats
popularity = pd.read_csv('./data/popularity.csv')
print(popularity.shape)
popularity.head()
```

```python
# there are multiple columns with space
# remove the space
popularity.rename(columns= lambda x: x.strip(), inplace=True)
popularity.head()
```

```python
#mode
print(stats.mode(popularity['num_keywords']))
#mean
print(np.mean(popularity['shares']))
#median
print(np.median(popularity['shares']))
#standard deviation
print(np.std(popularity['shares']))
```

```python
# All stats summary
popularity['shares'].describe()
```

```python
#percentile (check 78th percentile)
np.percentile(popularity['shares'],[78])
```

```python
# boxplot
plt.figure(figsize=(8,6))
plt.boxplot(x = popularity['shares'],widths = 0.6) # showfliers = False > remove outliers
plt.show()
```

```python
# Remove Outliers Using Normal Distribution and S.D.
# eliminating points that are above (Mean + 2*SD) and any points below (Mean - 2*SD) before plotting the frequencies.

#filter = 
np.mean(popularity['shares'])
```
